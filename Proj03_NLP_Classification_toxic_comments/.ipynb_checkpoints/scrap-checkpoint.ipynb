```python
#My code
#Function to sort out replies/snippets
comments=[]
def process_comments(response_items):
    
    for res in response_items:
        
        #Handle replies
        if 'replies' in res.keys():
            for reply in res['replies']['comments']:
                comment=reply['snippet']
                comment['comment_id']=reply['id']
                comments.append(comment)
        #Handle snippets
        else:
            comment={}
            comment['snippet']=res['snippet']['topLevelComment']['snippet']
            comment['snippet']['parent_id']=None #parent_id is set to None to keep the dataframe shape since there is no need to understand if a comment is a reply to another comment
            comment['snippet']['comment_id']=res['snippet']['topLevelComment']['id']
            comments.append(comment['snippet'])
    return comments

#Function to create a csv file from the dictionary extracted from API
import csv
def make_csv(comments, channelID):
    header=comments[0].keys()
    filename= f'comments_{channelID}.csv'
    
    with open(filename, 'w', encoding='utf8', newline='') as f:
        writer=csv.DictWriter(f, fieldnames=header, extrasaction='ignore')
        writer.writeheader()
        writer.writerows(comments) 
        
        
#Function to collect comments from Youtube's API

import os

from googleapiclient.discovery import build

youtube=build('youtube', 'v3',developerKey = 'AIzaSyB4RDwXJsoxS_PwdyZ-Q_NjzctaXFSlIbQ')

def comment_threads(channelID, to_csv=False):
    comments_list=[]
    
    request = youtube.commentThreads().list(
        part="id,replies,snippet",
        videoId=channelID
    )
    response = request.execute()
    comments_list.extend(process_comments(response['items']))
    
    while response.get('nextPageToken', None):
        request = youtube.commentThreads().list(
        part="id,replies,snippet",
        videoId=channelID,
        pageToken=response['nextPageToken']
    )
        response = request.execute()
        comments_list.extend(process_comments(response['items']))
    print(f"{len(comments_list)} number of comments found in video id: {channelID}")
    
    if to_csv:
        make_csv(comments_list, channelID)

def main(ID):
    comment_threads(ID, to_csv=True)
    print(f'csv file name: comments_{ID}.csv')
```
